```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
import numpy as np
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent = [
    'the bowl of cereal',
    'the cup of coffee',
    'the bottle of soda',
    'I am a dedicated learner',
    'I am a talented artist',
    'understand the importance of teamwork',
    'your ideas are creative',
    'the plate of cookies',
    'the jar of jam',
    'the cup of soup',
    'I am a passionate reader',
    'I am a skilled programmer',
    'remember the value of friendship',
    'your music is inspiring',
    'the slice of bread',
    'the glass of lemonade',
    'the bowl of fruit',
    'I am a curious explorer',
    'I am a supportive friend',
    'discover the joy of learning'
]
```

```python
print(sent)
```

```python
vocab_size=10000
```

```python
one_hot_repr=[one_hot(words,vocab_size)for words in sent]
one_hot_repr
```

```python
sent_length=10
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(vocab_size,dim,input_shape=(sent_length,)))
model.compile(optimizer='adam',loss='mse')
model.summary()
```

```python
model.predict(embedded_docs)
```
